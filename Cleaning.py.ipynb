```python
import pandas as pd
import numpy as np
```

```python
datos = pd.read_pickle('Datos/Data_all_sessions_cleaned')
```

```python
datos.head()
```

```python
file = open('Datos/Data_cleaned_to_MU','r')
change_to_MU = file.readlines()
file.close()
file = open('Datos/neuron_to_noise','r')
change_neuron_to_noise = file.readlines()
file.close()
file = open('Datos/noise_to_neuron','r')
change_noise_to_neuron = file.readlines()
file.close()
```

```python
names = {}
with open('names.txt','r') as f:
     for line in f:
            line = line.split(',')
            names[line[1].strip()] = line[0].strip()
```

```python
data = datos.copy()
for line in change_noise_to_neuron:
    try:
        line = line.strip()
        PE,ch,cl = line.split('_')
        data.loc[(data['PatientExperiment'] == names[PE]) & (data.Channel == int(ch) )& (data.Cluster == int(cl)),'bNoise'] = 0
    except:
        print('line = ',line)
for line in change_neuron_to_noise:
    try:
        line = line.strip()
        PE,ch,cl = line.split('_')
        data.loc[(data['PatientExperiment'] == names[PE]) & (data.Channel == int(ch) )& (data.Cluster == int(cl)),'bNoise'] = 1
    except:
        print('line = ',line)
for line in change_to_MU:
    try:
        line = line.strip()
        PE,ch,cl = line.split('_')
        data.loc[(data['PatientExperiment'] == names[PE]) & (data.Channel == int(ch) )& (data.Cluster == int(cl)),'bNoise'] = 2
    except:
        print('line = ',line)
```

```python
data.to_pickle('Datos/Data_all_sessions_cleaned2')
```
