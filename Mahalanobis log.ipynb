```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Mahalanobis_log',names=['neurons_train','mu_train','noise_train','neurons_test','mu_test','noise_test','neurons_as_noise','neurons_as_mu','hits_neuron','mu_as_noise','mu_as_neuron','hits_mu','hits_noise','misses_noise'])
df
```

```python
df['hits_noise_pc'] = df.hits_noise/df.noise_test*100
hits_noise_mean = df.hits_noise_pc.mean()
hits_noise_std = df.hits_noise_pc.std()
print('hits noise = ',hits_noise_mean,hits_noise_std)

df['hits_neuron_pc'] = df.hits_neuron/df.neurons_test*100
hits_neuron_mean = df.hits_neuron_pc.mean()
hits_neuron_std = df.hits_neuron_pc.std()
print('hits neuron = ',hits_neuron_mean,hits_neuron_std)

df['hits_mu_pc'] = df.hits_mu/df.mu_test*100
hits_mu_mean = df.hits_mu_pc.mean()
hits_mu_std = df.hits_mu_pc.std()
print('hits mu = ',hits_mu_mean,hits_mu_std)

df['mu_neuron_pc'] = df.mu_as_neuron/df.mu_test*100
mu_neuron_mean = df.mu_neuron_pc.mean()
mu_neuron_std = df.mu_neuron_pc.std()
print('missed mu as neuron = ', mu_neuron_mean,mu_neuron_std)

df['mu_noise_pc'] = df.mu_as_noise/df.mu_test*100
mu_noise_mean = df.mu_noise_pc.mean()
mu_noise_std = df.mu_noise_pc.std()
print('missed mu as noise = ',mu_noise_mean,mu_noise_std)

df['neu_mu_pc'] = df.neurons_as_mu/df.neurons_test*100
neu_mu_mean = df.neu_mu_pc.mean()
neu_mu_std = df.neu_mu_pc.std()
print('missed neuron as mu = ',neu_mu_mean,neu_mu_std)

df['neu_noise_pc'] = df.neurons_as_noise/df.neurons_test*100
neu_noise_mean = df.neu_noise_pc.mean()
neu_noise_std = df.neu_noise_pc.std()
print('missed neurons as noise = ',neu_noise_mean,neu_noise_std)

df['falses_positives_pc'] = (df.neurons_as_noise + df.mu_as_noise)/(df.neurons_as_noise + df.mu_as_noise + df.hits_noise)*100
falses_positives_mean = df.falses_positives_pc.mean()
falses_positives_std = df.falses_positives_pc.std()
print('falses positives = ',falses_positives_mean,falses_positives_std)
```
